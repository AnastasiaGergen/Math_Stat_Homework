№3 Самостоятельно запрограммируйте функции, позволяющие рассчитывать значение выборочной функции распределения для каждого:
- Наблюдения в выборке
- Значения в произвольном векторе

```python
def ecdf1(x):   #значение выборочной функции распределения для каждого наблюдения в выборке                                    
    return np.array([np.mean(x <= i) for i in x])   
x = chi2.rvs(size = 50, df = 5) 
ecdf1(x)
```

```python
chi2.cdf(x, df = 5)
```

```python
def ecdf2(x, z):   #значение выборочной функции распределения для каждого значения вектора                                 
    return np.array([np.mean(x <= i) for i in z])   
x = chi2.rvs(size = 50, df = 5) 
ecdf2(x, [0.5, 1, 1.2, 1.8, 2, 2.6, 3, 5])
```

№4. Самостоятельно запрограммируйте функции, позволяющие рассчитывать значение гистограммы для каждого:
- Наблюдения в выборке
- Значения в произвольном векторе

```python
def hist1(x, bin_n):
    a = np.zeros(len(x))
    x_1 = x - min(x)
    h = (max(x) - min(x)) / bin_n # ширина столбца
    for i in range(len(x)):
        z = x[i] - min(x)
        bin_z = math.floor(z / h) + 1 # номер столбца
        a[i] = len(x[(x_1 >= ((bin_z - 1) * h)) & (x_1 < (bin_z * h))]) / (bin_n * h)
    return a
x = chi2.rvs(size = 50, df = 5) 
print(hist1(x, 50))
```

```python
def hist2(x, bin_n, v):
    a = np.zeros(len(v))
    x_1 = x - min(x)
    h = (max(x) - min(x)) / bin_n # ширина столбца
    for i in range(len(v)):
        z = v[i] - min(x)
        bin_z = math.floor(z / h) + 1 # номер столбца
        a[i] = len(x[(x_1 >= ((bin_z - 1) * h)) & (x_1 < (bin_z * h))]) / (bin_n * h)
    return a
x = chi2.rvs(size = 50, df = 5) 
print(hist2(x, 50, [0.5, 1, 1.2, 1.8, 2, 2.6, 3, 5]))
```

№ 5. Самостоятельно запрограммируйте функции, позволяющие по выборке строить график:
- Выборочной функции распределения
- Гистограммы с произвольным числом столбиков

```python
# График выборочной функции распределения
def ecdf_plot(x):
    x = np.sort(x)                                                                                    
    F_x = ecdf1(x)                             
    plt.xlabel('x')                                         
    plt.ylabel('F(x)')                                      
    plt.plot(x, F_x, '--', markersize = 1, label = "CDF", color = "limegreen")
    seaborn.ecdfplot(x, label = "ECDF", color = "palevioletred")
    plt.legend();
x = chi2.rvs(size = 1000, df = 5)
ecdf_plot(x);
```

```python
def hist(x, bins):
    x = np.sort(x)                                 
    plt.xlabel('x')                                         
    plt.ylabel('f(x)')                                      
    seaborn.histplot(x, stat = 'density', color = "palevioletred", label = "histogram", bins = bins)                           
    plt.legend()
x = chi2.rvs(size = 1000, df = 5)
hist(x, 50);   
```

№6. Используя ЗБЧ и несколько выборок из Хи-Квадрат распределения приблизительно рассчитайте:
- Математическое ожидание выборочной функции распределения в точке 5
- Дисперсию выборочной функции распределения в точке 5

```python
m = 1000                                                
x_dist = np.empty(m)                                    
for i in range(0, m):
    x_new = chi2.rvs(size = 5000, df = 10)
    x_dist[i] = np.mean(x_new <= 5)
x_dist.mean()
```
